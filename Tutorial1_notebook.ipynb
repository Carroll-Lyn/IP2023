```python
img = cv2.imread('base.jpg',0)
cv2.imshow('base',img)
k = cv2.waitKey(0)
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('base.jpg',img)
    cv2.destroyAllWindows()
```
